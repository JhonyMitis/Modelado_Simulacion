```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def vpd(t,z):
  x,y = z
  return [y,mu*(1-x**2)*y-x]

a,b = 0,10
mus = [0,1,2,5]
styles = ['','-','.','*']

t = np.linspace(a,b,500)
sol = solve_ivp(vpd,[a,b],[1,0],t_eval= t)
plt.plot(sol.y[0],sol.y[1])
```

```
for mu,style in zip(mus,styles):
  sol = solve_ivp(vpd,[a,b],[1,0],t_eval= t)
  plt.plot(sol.t,sol.y[0],style)
  plt.axes().set_aspect(1)
```

```
def root(t,y): return y[0]-y[1]

mu = 2
sol = solve_ivp(vpd,[0,40],[1,0],t_eval=t,events=root)
zeros = sol.t_events[0]
space = zeros[1:]-zeros[:-1]
delta = zeros[1:]-zeros[:-1]
print(zeros)
plt.plot(space)
```
